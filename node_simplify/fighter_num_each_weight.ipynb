```python
# UFC 체급 표
# ------------------------------------------------
# strawweight <= 52.3kg
# 52.3kg < flyweight <= 56.7kg
# 56.7kg < bantamweight <= 61.2kg
# 61.2kg < featherweight <= 65.8kg
# 65.8kg < lightweight <= 70.3kg
# 70.3kg < welterweight <= 77.1kg
# 77.1kg < middleweight <= 83.9kg
# 83.9kg < light heavyweight <= 93.0kg
# 93.0kg < heavyweight <= 120.2kg
# -------------------------------------------------
```

```python
import pandas as pd
```

```python
# 파일 로드
file_path = 'fighter_stats.csv'  # 파일 경로
data = pd.read_csv(file_path)
```

```python
data.head(10)
```

```python
# weight 열에서 숫자가 아닌 값을 가진 데이터 필터링
non_numeric_weights = data[~data['weight'].str.replace('.', '', 1).str.isdigit()]

# 출력
print("weight 열에 숫자가 아닌 값을 가진 데이터:")
print(non_numeric_weights)
```

```python
# 2470번 행부터 2480번 행까지 데이터 확인
subset_data = data.loc[2470:2480]

# 확인 출력
print(subset_data)
```

```python
# 2478번 행 제거
data = data.drop(index=2478)

# 제거 후 데이터프레임 확인
print(data.loc[2470:2480])
```

```python
# weight 열에서 숫자가 아닌 값을 가진 데이터 필터링
non_numeric_weights = data[~data['weight'].str.replace('.', '', 1).str.isdigit()]

# 출력
print("weight 열에 숫자가 아닌 값을 가진 데이터:")
print(non_numeric_weights)
```

```python
# 숫자로 변환할 수 없는 데이터 확인
non_numeric_weights = data[~data['weight'].apply(lambda x: isinstance(x, (int, float)) or str(x).replace('.', '', 1).isdigit())]
print("Non-numeric weights data:")
print(non_numeric_weights)
```

```python
# weight 열을 숫자로 변환 (변환 불가능한 값은 NaN 처리)
data['weight'] = pd.to_numeric(data['weight'], errors='coerce')
```

```python
# 체급 기준에 따라 체급 분류
def categorize_weight(weight):
    if weight <= 52.3:
        return 'Strawweight'
    elif weight <= 56.7:
        return 'Flyweight'
    elif weight <= 61.2:
        return 'Bantamweight'
    elif weight <= 65.8:
        return 'Featherweight'
    elif weight <= 70.3:
        return 'Lightweight'
    elif weight <= 77.1:
        return 'Welterweight'
    elif weight <= 83.9:
        return 'Middleweight'
    elif weight <= 93.0:
        return 'Light Heavyweight'
    elif weight <= 120.2:
        return 'Heavyweight'
    else:
        return 'Unknown'  # 체급 범위를 벗어난 경우
```

```python
# 체급 컬럼 생성
data['Weight Class'] = data['weight'].apply(categorize_weight)
```

```python
# 체급별 선수 수 계산
weight_class_counts = data['Weight Class'].value_counts()
```

```python
# 출력
print("체급별 선수 수:")
print(weight_class_counts)
```

```python
# 체급이 'Unknown'인 선수들의 이름 확인
unknown_fighters = data[data['Weight Class'] == 'Unknown']['name']

# 결과 출력
print("Unknown 체급에 포함된 선수들:")
print(unknown_fighters)
```

```python
# unknown에 포함된 선수들은 어떤 선수들인가?
# Emmanuel Yarborough ~ Mark Kerr 조사결과, 00년대 이전의 ufc 선수들
# 300kg가 넘어가는 사람도 있었다.
# 체급이란 개념이 정리되기 이전의 선수들인 것 같아 제외하기로 함
```

```python
# Unknown 체급 데이터 제거
data = data[data['Weight Class'] != 'Unknown']

# 결과 확인
print("Unknown 체급이 제거된 데이터셋입니다.")
print(data['Weight Class'].value_counts())
```
