```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import torch
from torch_geometric.data import Data, DataLoader
from torch_geometric.nn import GCNConv
import torch.nn.functional as F
```

```python
# Load the dataset
data = pd.read_csv("large_dataset.csv")
```

```python
# Preprocessing: Fill missing values and convert categorical data
data.fillna(value={"r_height": data["r_height"].median(),
                   "b_height": data["b_height"].median(),
                   "r_weight": data["r_weight"].median(),
                   "b_weight": data["b_weight"].median()}, inplace=True)

categorical_columns = ["gender", "method", "r_stance", "b_stance", "weight_class"]
data = pd.get_dummies(data, columns=categorical_columns, drop_first=True)
```

```python
# Feature selection
features = [
    "r_kd", "r_sig_str", "r_sig_str_att", "r_td", "r_td_att",
    "b_kd", "b_sig_str", "b_sig_str_att", "b_td", "b_td_att",
    "r_height", "r_weight", "b_height", "b_weight"
]
X = data[features]
y = data["winner"].apply(lambda x: 1 if x == "Red" else 0)
```

```python
# Scaling the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Fighter Clustering
kmeans = KMeans(n_clusters=10, random_state=42)  # Adjust clusters as needed
data["fighter_cluster"] = kmeans.fit_predict(X_scaled)
```

```python
# 모든 선수 이름 추출
all_fighters = set(data['r_fighter'].dropna().unique()).union(
    set(data['b_fighter'].dropna().unique())
)

# 선수 이름을 고유한 정수 인덱스로 매핑
fighter_to_node = {fighter: idx for idx, fighter in enumerate(all_fighters)}
```

```python
# Graph Construction
def construct_graph(data):
    edge_index = []
    edge_attr = []
    outcomes = []

    for index, row in data.iterrows():
        r_fighter = row['r_fighter']
        b_fighter = row['b_fighter']
        outcome = 1 if row['winner'] == 'Red' else 0  # 승리 결과를 숫자로 변환

        if pd.isna(r_fighter) or pd.isna(b_fighter):  # 결측값 처리
            continue

        if not isinstance(outcome, (int, float)):  # 숫자가 아닌 값 처리
            continue

        r_index = fighter_to_node[r_fighter]
        b_index = fighter_to_node[b_fighter]
        edge_index.append([r_index, b_index])
        edge_attr.append([row['total_rounds'], row['finish_round']])
        outcomes.append(outcome)

    # Tensor로 변환
    edge_index = torch.tensor(edge_index, dtype=torch.long).t().contiguous()
    edge_attr = torch.tensor(edge_attr, dtype=torch.float)
    outcomes = torch.tensor(outcomes, dtype=torch.float)

    return edge_index, edge_attr, outcomes

edge_index, edge_attr, outcomes = construct_graph(data)
```

```python
# Graph Neural Network
class FightPredictor(torch.nn.Module):
    def __init__(self, input_dim, hidden_dim, output_dim):
        super(FightPredictor, self).__init__()
        self.conv1 = GCNConv(input_dim, hidden_dim)
        self.conv2 = GCNConv(hidden_dim, output_dim)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)
```

```python
# Prepare data for PyTorch Geometric
graph_data = Data(
    x=torch.tensor(X_scaled, dtype=torch.float),
    edge_index=edge_index,
    edge_attr=edge_attr,
    y=outcomes
)
```

```python
# DataLoader
loader = DataLoader([graph_data], batch_size=1, shuffle=True)
```

```python
# Model Training
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = FightPredictor(input_dim=len(features), hidden_dim=32, output_dim=2).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
loss_fn = torch.nn.CrossEntropyLoss()

for epoch in range(2000):  # Adjust the number of epochs as needed
    model.train()
    for batch in loader:
        batch = batch.to(device)
        optimizer.zero_grad()
        out = model(batch.x, batch.edge_index)
        loss = loss_fn(out, batch.y.long())
        loss.backward()
        optimizer.step()
    print(f"Epoch {epoch+1}, Loss: {loss.item()}")
```

```python
# Save the model
torch.save(model.state_dict(), "ufc_gnn_model.pt")
```
